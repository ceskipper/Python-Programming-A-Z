```python
#FieldGoals per Game
FieldGoalsperGame = []

for i in range(0, len(FieldGoals)):       #number of rows aka length of the rows
    temp = []
    for j in range(0, len(FieldGoals[i])):
        if Games[i][j] != 0:
            temp.append(round(FieldGoals[i][j]/Games[i][j]))
        else:
            temp.append(0)
    #temp = round(FieldGoals[i]/Games[i])
    #print(temp)
    FieldGoalsperGame.append(temp)

print(np.array(FieldGoalsperGame))
```

```python
#FieldGoals per Game
import warnings
warnings.filterwarnings('ignore')
FieldGoalsPerGame = np.matrix.round(FieldGoals / Games)
print(FieldGoalsPerGame)
```

#### Slice using :
#### Call using ,

```python
import numpy as np
A = np.array([[1,2,3,], [10,20,30],[100,200,300]])
print(A)
```

```python
A[:,2]   #slice all of col 2
```

```python
A[2,:]   #slice all of row 2
```

```python
A[2,2]   #row 2, col 2
```

```python
A[2]     #call row 2
```

```python
A[,2]    #doesn't work! Must specify all of the rows using :
```
