### Three Parts to Subset DFs:
#### -Subsetting Rows
#### -Subsetting Columns
#### -Combining Rows & Columns

### 1) Subsetting Rows
#### Because dataframes and tables are most often organized with the data in rows (instead of columns), Python automatically assumes you want to slice by rows

```python
stats[21:26]             #in the [] specify which rows you want to subset; does NOT include row 26 (aka the last row)
```

```python
stats[:26]                #will subset rows 0 to 26
```

```python
stats[:10]   #same as stats.head(10)
```

## Using .at() and .iat() - advanced tutorial
### Accessing individual elements
#### .at()    for labels. Important: even integers are treated as labels
#### .iat()   for integer location

```python
#stats[2,2]   #error. why?!
```

```python
stats.head()
```

### .iat()       counts using integers, so it counts the rows and columns you specify

```python
#stats[3,4]   #error!
```

```python
stats.iat[3,4]   #row 3, column 4
```

```python
stats.iat[0,1]
```

### .at()   locates using labels, so you would include the column name

```python
stats.at[2,'BirthRate']
```

#### why we need .at():

```python
sub10 = stats[::10]
```

```python
sub10
```

```python
sub10.iat[10,0]    #treats 10 as the row number, NOT as a label (like .at would)
```

```python
sub10.at[10,'CountryName'] #treats 10 (an integer) as a label instead of as row 10!
```
